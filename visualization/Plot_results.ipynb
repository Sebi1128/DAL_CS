## Load the results from the csv files

```python
def load_csv(filepath):
    data_list = []
    with open(filepath, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quoting=csv.QUOTE_NONNUMERIC)
        for row in reader:
            data_list.append(row)
    return np.asarray([item for sublist in data_list for item in sublist])
```

```python
mnist_runs_data = {}
for key in mnist_runs_id:
    mnist_path = save_path + 'MNIST/' + key + '.csv'
    mnist_runs_data[key] = load_csv(mnist_path)

fmnist_runs_data = {}
for key in fmnist_runs_id:
    fmnist_path = save_path + 'FMNIST/' + key + '.csv'
    fmnist_runs_data[key] = load_csv(fmnist_path)
    
cifar10_runs_data = {}
for key in cifar10_runs_id:
    cifar10_path = save_path + 'CIFAR10/' + key + '.csv'
    cifar10_runs_data[key] = load_csv(cifar10_path)
    
cifar100_runs_data = {}
for key in cifar100_runs_id:
    cifar100_path = save_path + 'CIFAR100/' + key + '.csv'
    cifar100_runs_data[key] = load_csv(cifar100_path)
    
cifar10_params_runs_data = {}
for key in params_runs_id:
    cifar10_params_path = save_path + 'PARAMS/' + key + '.csv'
    cifar10_params_runs_data[key] = load_csv(cifar10_params_path)
    
cifar10_ablation_runs_data = {}
for key in cifar10_ablation_runs_id:
    cifar10_ablation_path = save_path + 'ABLATION/' + key + '.csv'
    cifar10_ablation_runs_data[key] = load_csv(cifar10_ablation_path)
```

## Plot the results

### MNIST accuracy

```python
# Compute mean and standard deviation for MNIST random sampling
mnist_rnd = np.stack((mnist_runs_data["mnist_rnd_42"],
                      mnist_runs_data["mnist_rnd_26"],
                      mnist_runs_data["mnist_rnd_15"]),
                     axis=0)
mnist_rnd_mean = np.mean(mnist_rnd, axis=0)
mnist_rnd_std = np.std(mnist_rnd, axis=0)

# Compute mean and standard deviation for MNIST CAL sampling
mnist_cal = np.stack((mnist_runs_data["mnist_cal_42"],
                      mnist_runs_data["mnist_cal_26"],
                      mnist_runs_data["mnist_cal_15"]),
                     axis=0)
mnist_cal_mean = np.mean(mnist_cal, axis=0)
mnist_cal_std = np.std(mnist_cal, axis=0)


# Sampling ratios
ratios = [100.*item for item in [0.02, 0.04, 0.06, 0.08, 0.1]]


fig, ax = plt.subplots(dpi=500)
plt_props()


mnist_cal_plot = plt.errorbar(ratios, mnist_cal_mean, label='MNIST CAL', color='b' ,yerr=mnist_cal_std)
mnist_rnd_plot = plt.errorbar(ratios, mnist_rnd_mean, label='MNIST random', color='g' ,yerr=mnist_rnd_std)

plt.legend(ncol=1, loc='lower right')
plt.xlabel('% of Labeled Data')
plt.ylabel('Mean Accuracy (%)')  
plt.title('MNIST')
plt.grid(True)
plt.show()
fig.savefig(save_path_plots+'acc_mnist.pdf', dpi=500)
```

### Ablation study

```python
# Compute mean and standard deviation for CIFAR10 ablation: PCA latent space
cifar10_abl_pca = np.stack((cifar10_ablation_runs_data["cifar10_ablation_pca32_clsVAAL_smpCAL_42"],
                            cifar10_ablation_runs_data["cifar10_ablation_pca32_clsVAAL_smpCAL_26"],
                            cifar10_ablation_runs_data["cifar10_ablation_pca32_clsVAAL_smpCAL_15"]),
                           axis=0)
cifar10_abl_pca_mean = np.mean(cifar10_abl_pca, axis=0)
cifar10_abl_pca_std = np.std(cifar10_abl_pca, axis=0)

# Compute mean and standard deviation for CIFAR10 ablation: CAL sampling with KL-divergence
cifar10_abl_kl = np.stack((cifar10_ablation_runs_data["cifar10_ablation_vae32_clsVAAL_smpCAL(kl)_42"],
                           cifar10_ablation_runs_data["cifar10_ablation_vae32_clsVAAL_smpCAL(kl)_26"],
                           cifar10_ablation_runs_data["cifar10_ablation_vae32_clsVAAL_smpCAL(kl)_15"]),
                          axis=0)
cifar10_abl_kl_mean = np.mean(cifar10_abl_kl, axis=0)
cifar10_abl_kl_std = np.std(cifar10_abl_kl, axis=0)

# Compute mean and standard deviation for CIFAR10 ablation: Classifier with latent space
cifar10_abl_latent = np.stack((cifar10_ablation_runs_data["cifar10_ablation_vae32_clsVAALwZ_smpCAL_42"],
                               cifar10_ablation_runs_data["cifar10_ablation_vae32_clsVAALwZ_smpCAL_26"],
                               cifar10_ablation_runs_data["cifar10_ablation_vae32_clsVAALwZ_smpCAL_15"]),
                              axis=0)
cifar10_abl_latent_mean = np.mean(cifar10_abl_latent, axis=0)
cifar10_abl_latent_std = np.std(cifar10_abl_latent, axis=0)


# Sampling ratios
ratios = [100.*item for item in [0.1, 0.15, 0.2, 0.25, 0.3]]

fig, ax = plt.subplots(dpi=500)
plt_props()

# Compared to standard model
cifar10_cal_plot = plt.errorbar(ratios, cifar10_cal_mean, label='CAL (no ablation)', color='b' ,yerr=cifar10_cal_std)

# CIFAR10 ablation runs
cifar10_abl_kl_plot = plt.errorbar(ratios, cifar10_abl_kl_mean, label='CAL with KL divergence', color='m' ,yerr=cifar10_abl_kl_std)
cifar10_abl_latent_plot = plt.errorbar(ratios, cifar10_abl_latent_mean, label='Classifier on image and latent space', color='darkred' ,yerr=cifar10_abl_latent_std)
cifar10_abl_pca_plot = plt.errorbar(ratios, cifar10_abl_pca_mean, label='PCA to construct latent space', color='c' ,yerr=cifar10_abl_pca_std)


plt.legend(ncol=1, loc='lower right')
plt.xlabel('% of Labeled Data')
plt.ylabel('Mean Accuracy (%)')  
plt.title('Ablation on CIFAR10')
plt.grid(True)
plt.show()
fig.savefig(save_path_plots+'ablation_study.pdf', dpi=500)
```
